```python
# importing required libraries
import seaborn as sns
sns.set()
sns.set(style="darkgrid")


import numpy as np
import pandas as pd

# importing matplotlib
import matplotlib.pyplot as plt
%matplotlib inline

import warnings
warnings.filterwarnings("ignore")
plt.rcParams['figure.figsize']=(10,10)
```

```python
# read the dataset
data_BM = pd.read_csv('https://raw.githubusercontent.com/mohdsanadzakirizvi/Matplotlib-Seaborn-Tutorial/master/Data_Visualization_with_Seaborn/bigmart_data.csv')

# drop the null values
data_BM = data_BM.dropna(how="any")

# multiply Item_Visibility by 50 to increase size
data_BM["Visibility_Scaled"] = data_BM["Item_Visibility"] * 50

# view the top results
data_BM.head()
```

```python
# line plot using relplot
sns.lineplot(x="Outlet_Establishment_Year", y="Item_MRP",data=data_BM[:100]);
```

```python
# Bar plot
sns.barplot(x="Item_Type", y="Item_MRP", data=data_BM[:5])
```

```python
# Histogram
sns.distplot(data_BM['Item_Outlet_Sales'])
```

```python
# Box plot
sns.boxplot(data_BM['Item_Weight'], orient='h')
```

```python
#Violin plot
sns.violinplot(data_BM['Item_Outlet_Sales'], orient='vertical', color='green')
```

```python
# scatter plot
sns.relplot(x="Item_MRP", y="Item_Outlet_Sales", data=data_BM[:100], kind="scatter");
```

```python
#Hue semantic
sns.relplot(x="Item_MRP", y="Item_Outlet_Sales", hue="Item_Type",data=data_BM[:100]);
```

```python
# Different line plots for different categories of the Outlet_Size
sns.lineplot(x="Item_Weight", y="Item_MRP",hue='Outlet_Size',data=data_BM[:30]);
```

```python
# Bubble plot
sns.relplot(x="Item_MRP", y="Item_Outlet_Sales", data=data_BM[:100], kind="scatter", size="Visibility_Scaled", hue="Visibility_Scaled");
```

```python
# Subplots for each of the category of Outlet_Size
sns.relplot(x="Item_Weight", y="Item_Visibility",hue='Outlet_Size',style='Outlet_Size',col='Outlet_Size',data=data_BM[:100]);
```

```python
# Strip plot
sns.catplot(x="Outlet_Size", y="Item_Outlet_Sales", kind='strip',data=data_BM[:250]);
```

```python
# Swarm plot
sns.catplot(x="Outlet_Size", y="Item_Outlet_Sales", kind='swarm',data=data_BM[:250]);
```

```python
# Box Plots
sns.catplot(x="Outlet_Size", y="Item_Outlet_Sales",kind="box",data=data_BM);
```

```python
# Violin Plots
sns.catplot(x="Outlet_Size", y="Item_Outlet_Sales",kind="violin",data=data_BM);
```

```python
# Point plot
sns.catplot(x="Outlet_Size", y="Item_Outlet_Sales",kind="point",data=data_BM);
```

```python
# Bar plots
sns.catplot(x="Outlet_Size", y="Item_Outlet_Sales",kind="bar",data=data_BM);
```

```python
# distribution of Item_MRP
plt.figure(figsize=(10,10))
sns.kdeplot(data_BM['Item_MRP'], shade=True);
```

```python
# Histogram and Density Plot
plt.figure(figsize=(10,10))
sns.distplot(data_BM['Item_Outlet_Sales']);
```
